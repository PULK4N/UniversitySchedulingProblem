```python
def plot_over_time(data):
  data_desc = data.groupby(by='generation').agg({'fitness_score': ['mean', 'min', 'max']})
  fig = plt.figure()
  plt.xlabel("generation")
  plt.ylabel("fitness score")
  plt.plot(data_desc[('fitness_score', 'min')], '^b')
  plt.plot(data_desc[('fitness_score', 'mean')], '^g')
  plt.plot(data_desc[('fitness_score', 'max')] ,'^r')
  plt.title('Fitness scores over generations')
  plt.legend(['min','mean','max'])
```
