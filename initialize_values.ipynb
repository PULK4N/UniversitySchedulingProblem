```python
class School_subject_session:
    id = 0
    def __init__(self,subject_id,group_id, name,student_group,professor,start_end_time,classroom = 0):
        self.id = int(str(subject_id)+str(group_id))
        self.name = name
        self.student_group = student_group
        self.professor = professor
        self.start_time = datetime.datetime(2022, 1, 1 + int(start_end_time[0]/24), start_end_time[0] % 24)
        self.end_time = datetime.datetime(2022, 1, 1 + int(start_end_time[0]/24), start_end_time[1] % 24)
        self.classroom = classroom
        
    def __str__(self) -> str:
        return 'Student group: {}, professor: {}, from {} untill {}, classroom: {}'.format(
            self.student_group,self.professor,self.start_time,self.end_time,self.classroom
        )
```

```python
class School_subject:
    id = 0
    def __init__(self,name,duration = 1):
        School_subject.id += 1
        self.id = School_subject.id
        self.name = name
        self.student_groups = []
        self.professors = []
        self.begin_end_times = self.generate_begin_end_times(duration)

    def __str__(self) -> str:
        school_subject_to_str = 'Professors:\n'
        for professor in self.professors:
            school_subject_to_str += str(professor) + '\n'
        school_subject_to_str += 'Student groups:\n'
        for student_group in self.student_groups:
            school_subject_to_str += str(student_group) + '\n'

        return "id: {}, Name: {}\n{}".format(self.id,self.name,school_subject_to_str)
    #create all possible class time sessions by given duration of the class
    def generate_begin_end_times(self,duration):
        begin_end_times = []
        for j in range(0,5):
            for i in range(8,20-duration):
                begin_end_times.append((j*24 + i,j*24 + i+duration))
        return begin_end_times
```
